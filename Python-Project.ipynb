```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the data set into Dataframe

data = pd.read_excel('Dataset.xlsx')
data.head(20)
```

# Preprocessing

```python
# Replace the 'Height' column with random numbers between 150 and 180.

data['Height'] = np.random.randint(150,181, size=len(data))
```

```python
data.head(10)
```

```python
data.isnull().sum()
```

```python
data['Salary'].mean()
```

```python
data['Salary'].median()
```

```python
# Fill missing values in "Salary" with the median salary

data.fillna({'Salary':data['Salary'].median()}, inplace=True)
```

```python
# Fill missing values in "College" with 'Unknown'

data.fillna({'College':'Unknown'}, inplace=True)
```

```python
data.isnull().sum()
```

```python
data.head(10)
```

# Analysis Tasks

### 1. Employee Distribution by Team

```python
data
```

```python
# The number of employees in each team and the percentage split

emp_count = data['Team'].value_counts()
percentage = (emp_count / len(data)) * 100

team_distribution = pd.DataFrame({'Employee count':emp_count,'Percentage':percentage})
team_distribution
```

```python
# Graphical representation

plt.figure(figsize=(12, 5))
sns.barplot(x=team_distribution.index, y=team_distribution["Employee count"], color='#647a02')
plt.xlabel("Team")
plt.ylabel("Employee Count")
plt.title('Employee Distribution Across Teams')
plt.xticks(rotation=90)
plt.show()
```

### 2. Employee Segregation by Position

```python
data
```

```python
# Distribution of employees in each position
position_counts = data["Position"].value_counts()

percentage = (position_counts / len(data)) * 100

position_distribution = pd.DataFrame({"Employee Count": position_counts, "Percentage": percentage})
position_distribution
```

```python
# Visualization

plt.figure(figsize=(7, 7))
plt.pie(position_distribution["Percentage"], labels=position_distribution.index, autopct="%1.1f%%")
plt.title("Distribution of Employees by Position")
plt.show()
```

### 3. Predominant Age Group

```python
# Define age groups
bins = [18, 25, 30, 35, 40, 50]  # Age ranges
labels = ["18-24", "25-29", "30-34", "35-39", "40-49"]

# Create a new column for age groups
data["Age Group"] = pd.cut(data["Age"], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = data["Age Group"].value_counts()

pd.DataFrame(age_group_counts)
```

```python
#Visualization

plt.figure(figsize=(8, 5))
sns.countplot(data=data, x='Age Group', color='indigo')
plt.xlabel("Age Group")
plt.ylabel("Employees Count")
plt.title("Employee Distribution Across Age Groups")
plt.show()
```

### 4. Highest Salary Expenditure by Team & Position

```python
#Team salary expenditure

team_salary = data.groupby("Team")["Salary"].sum().sort_values(ascending=False)
df1 = pd.DataFrame(team_salary)
df1
```

```python
print('Team with highest salary expenditure:\n',team_salary.head(1))
```

```python
# Visualization

plt.figure(figsize=(12, 5))
sns.barplot(x=df1.index, y=df1['Salary'], color='#e33a68')
plt.xlabel("Team")
plt.ylabel("Salary Expenditure")
plt.title('Total Salary Expenditure by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
#Position salary expenditure

position_salary = data.groupby("Position")["Salary"].sum().sort_values(ascending=False)
df2 = pd.DataFrame(position_salary)
df2
```

```python
print('Position with highest salary expenditure:\n',position_salary.head(1))
```

```python
# Visualization

plt.figure(figsize=(8, 5))
sns.barplot(x=df2.index, y=df2['Salary'], color='#fd9e14')
plt.xlabel("Position")
plt.ylabel("Salary Expenditure")
plt.title('Total Salary Expenditure by Position')
plt.xticks(rotation=90)
plt.show()
```

### 5. Correlation Between Age & Salary

```python
# correlation between Age and Salary

correlation = data["Age"].corr(data["Salary"])
print('correlation between age and salary:',correlation)
```

```python
# Visualization

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
sns.regplot(x=data["Age"], y=data["Salary"], scatter=False, color="red")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title('Correlation between Age and Salary')
plt.grid()
plt.show()
```

# Data Story

####  Employee Distribution Trends:
 New Orleans Pelicans have the highest number of employees (4.15% of total).  
 Most teams have a fairly even distribution of employees (around 15 per team).

#### Segregation by Position:
 The most common position is Shooting Guard (SG) (22.27% of total employees).  
 Centers (C) are the least common, making up only 17.25% of employees.

#### Age Group Analysis:
 25-29 years is the predominant age group (39.74% of employees).  
Employees above 40 years are very rare (only 0.66% of total workforce).

#### Salary Expenditure Insights:
 Cleveland Cavaliers have the highest total salary expenditure (106,988,689).  
 The Center (C) position has the highest total salary expenditure ($466,377,332).

#### Correlation Between Age and Salary:
There is a weak positive correlation (r = 0.21) between Age and Salary.  
Older employees tend to have slightly higher salaries, but the trend is not strong.
