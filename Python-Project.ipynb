##### 1.The distribution of employees across each team and the percentage split relative to the total number of employees.

```python
#calculate the number of employees in each team
emp_count = data['Team'].value_counts()

#calculate the percentage split
percentage = (emp_count / len(data)) * 100

#to display the result
team_distribution = pd.DataFrame({'Employee count':emp_count,'Percentage':percentage})
team_distribution
```

##### 3. Identify the predominant age group among employees.

```python
# Define age groups
bins = [18, 25, 30, 35, 40, 50]  # Age ranges
labels = ["18-24", "25-29", "30-34", "35-39", "40-49"]

# Create a new column for age groups
data["Age Group"] = pd.cut(data["Age"], bins=bins, labels=labels, right=False)

# Count the number of employees in each age group
age_group_counts = data["Age Group"].value_counts()

pd.DataFrame(age_group_counts)
```

##### 4. Discover which team and position have the highest salary expenditure

```python
#Team salary expenditure

team_salary = data.groupby("Team")["Salary"].sum().sort_values(ascending=False)
df1 = pd.DataFrame(team_salary)
df1
```

```python
print('Team with highest salary expenditure:\n',team_salary.head(1))
```

```python
# Visualization

plt.figure(figsize=(12, 5))
sns.barplot(x=df1.index, y=df1['Salary'], color='brown')
plt.xlabel("Team")
plt.ylabel("Salary Expenditure")
plt.title('Total Salary Expenditure by Team')
plt.xticks(rotation=90)
plt.show()
```

```python
#Position salary expenditure

position_salary = data.groupby("Position")["Salary"].sum().sort_values(ascending=False)
df2 = pd.DataFrame(position_salary)
df2
```

```python
print('Position with highest salary expenditure:\n',position_salary.head(1))
```

```python
# Visualization

plt.figure(figsize=(12, 5))
sns.barplot(x=df2.index, y=df2['Salary'], color='pink')
plt.xlabel("Position")
plt.ylabel("Salary Expenditure")
plt.title('Total Salary Expenditure by Position')
plt.xticks(rotation=90)
plt.show()
```

##### 5. Investigate if there's any correlation between age and salary, and represent it visually

```python
# correlation between Age and Salary

correlation = data["Age"].corr(data["Salary"])
print('correlation between age and salary:',correlation)
```

```python
# Visualization

plt.figure(figsize=(10, 6))
sns.scatterplot(x=data['Age'], y=data['Salary'], color='purple')
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title('Correlation between Age and Salary')
plt.grid()
plt.show()
```
